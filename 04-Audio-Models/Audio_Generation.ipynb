```python
import transformers
```

```python
from transformers import pipeline
pipe = pipeline("text-to-speech")
```

```python
text = "Hellow, how are you toda?"
```

```python
output = pipe(text)
```

```python
output['audio']
```

```python
output['sampling_rate']
```

```python
from IPython.display import Audio
Audio(output['audio'], rate=output['sampling_rate'])
```

```python
from pydub import AudioSegment
```

```python
output['audio']
```

```python
audio_bytes = output['audio'].tobytes()
audio_seg = AudioSegment(audio_bytes,
                         frame_rate=output['sampling_rate'],
                         sample_width = output['audio'].dtype.itemsize,
                         channels = 1)
```

```python
audio_seg.export("tts_example.mp3", format="mp3")
```
